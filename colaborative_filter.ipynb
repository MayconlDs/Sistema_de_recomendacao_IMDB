Marge do dataset filme e rating 
- os dois datasets tem uma coluna de identifação em comum no caso movieId.

```python
df = filme.merge(rating, on='movieId')
```

```python
df.head()
```

Pivot Table - filtro colaborativo
- nessa parte do estudo iremos cruzar as informações dos filmes x usuarios juntamento com a nota que cada um,
atribuiu ao filme assitido por eles.

```python
tabela_dados = pd.pivot_table(df, index='title', columns='userId', values='rating')
tabela_dados
```

```python
#não podemos deixar a tabela com valores NaN pois o algoritmo espera números para poder performar 
#então usaremos um recurso simples para preencher os valores faltantes com o comando fillna(0)
tabela_dados_modificados = pd.pivot_table(df, index='title', columns='userId', values='rating').fillna(0)
tabela_dados_modificados
```

```python
recomendacao = pw.cosine_similarity(tabela_dados_modificados)
recomendacao_df = pd.DataFrame(recomendacao, columns=tabela_dados.index, index=tabela_dados.index)
```

```python
recomendacao_df.head()
```

```python
def sistema_rec(texto):
    sistema_recomendacao = pd.DataFrame(recomendacao_df[f"{texto}"].sort_values(ascending=False))
    sistema_recomendacao.columns = ['Recomendações']
    return sistema_recomendacao.head(10)[1:].reset_index(names=['Titulo'])
```
